# Exercise 4

```r
library(e1071)
set.seed(1)
x <- rnorm(100)
y <- 3 * x^2 + 2 + rnorm(100)
class <- sample(100, 30)
y[class] <- y[class] + 3
y[-class] <- y[-class] - 3
plot(x[class], y[class], col = "red", xlab = "X", ylab = "Y", ylim = c(-6, 30))
points(x[-class], y[-class], col = "yellow")
```

# Exercise 8

```r
library(ISLR)
set.seed(1)
train <- sample(nrow(OJ), 800)
df_train <- OJ[train, ]
df_test <- OJ[-train, ]
```

```r
svm <- svm(Purchase ~ ., data = df_train, kernel = "linear", cost = 0.01)
summary(svm)
```

```r
pred <- predict(svm, df_train)
table(df_train$Purchase, pred)
```

```r
pred_test <- predict(svm, df_test)
table(df_test$Purchase, pred_test)
```
