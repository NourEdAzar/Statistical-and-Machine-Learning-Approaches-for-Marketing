```r
install.packages("ISLR")
library(ISLR)
```

```r
head(Carseats)
```

```r
#a#
#Encoding Urban variable
Carseats$Urban <- ifelse(Carseats$Urban == 'Yes', 1,0)
#Encoding US variable
Carseats$US <- ifelse(Carseats$US == 'Yes', 1,0)

lm.fit = lm(Sales ~ Price + Urban + US, data = Carseats)
summary(lm.fit)
```

```r
#b#
# there is an inverse relation between price and sales and Urban and sales, but the relationship is not strong.
# however there is a strong positive relationship between the Sales and US.

#c#
# Sales = 13.04347 - 0.05446 Price - 0.02192 Urban + 1.20057 USD + ε

#d#
# the p value of Price and US is <0.05 so we reject the null hypothesis for them

#e#
lm.fit2 = lm(Sales ~ Price + US, data = Carseats)
summary(lm.fit2)

#f#
# They fit well since the r-squared is 0.2393

#g#
confint(lm.fit2)

#h#
plot(lm.fit2)
# The last graph shows there are outliers.

```
